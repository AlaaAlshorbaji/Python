# **3. Key Pandas Data Structures**

## **3.1 Series**

A Series is a one-dimensional array with labels (index).

```python
# Create a Series
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print(s)

# Attributes
print("Index:", s.index)
print("Values:", s.values)

# Accessing Data
print(s['a'])  # Access by label
print(s[0])    # Access by position
```

## **3.2 DataFrame**

A DataFrame is a two-dimensional labeled data structure.

```python
# Create a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'Salary': [50000, 60000, 70000]}
df = pd.DataFrame(data)

# Display DataFrame
print(df)

# Attributes
print("Columns:", df.columns)
print("Index:", df.index)
print("Shape:", df.shape)
```

## **3.3 Index Objects**

Pandas Index objects store the axis labels.

```python
# Access index
print(df.index)

# Set a new index
df.set_index('Name', inplace=True)
print(df)
```

# **6. Data Cleaning and Preprocessing**

## **6.1 Handling Missing Data**

```python
# Fill missing values
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Drop rows with missing values
df.dropna(inplace=True)
```

## **6.2 Handling Duplicates**

```python
# Drop duplicates
df.drop_duplicates(inplace=True)
```

# **7. Indexing, Selection, and Slicing**

Pandas offers various methods for selecting and slicing data.

```python
# Select a single column
print(df['Age'])

# Select multiple columns
print(df[['Name', 'Salary']])

# Slicing rows
print(df[0:3])

# Boolean indexing
print(df[df['Salary'] > 55000])
```

# **8. Advanced Operations on DataFrames**

## **8.1 Applying Functions**

```python
# Apply a function to a column
df['Bonus'] = df['Salary'].apply(lambda x: x * 0.1)

# Vectorized operations
df['Total Compensation'] = df['Salary'] + df['Bonus']
```

## **8.2 Sorting**

```python
# Sort by column
df.sort_values(by='Age', ascending=False, inplace=True)
```

## **8.3 Filtering**

```python
# Filter rows
filtered = df[(df['Age'] > 30) & (df['Salary'] > 55000)]
print(filtered)
```

# **9. Data Aggregation and Grouping**

```python
# Group by
grouped = df.groupby('Age')['Salary'].mean()
print(grouped)

# Multiple aggregations
aggregated = df.groupby('Age').agg({'Salary': ['mean', 'max'], 'Bonus': 'sum'})
print(aggregated)
```

# **10. Merging, Joining, and Concatenating Data**

Merge DataFrames

The pd.merge() function is used to combine two DataFrames based on a common column (or index). The on parameter specifies the column(s) to join on. This operation is similar to SQL joins, and you can choose the type of join (inner, outer, left, or right).

```python
import pandas as pd

# Sample data
df1 = pd.DataFrame({
    'Key': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'Key': [2, 3, 4],
    'Age': [30, 40, 50]
})

# Merge on the 'Key' column
merged = pd.merge(df1, df2, on='Key')
print(merged)
```

Concatenate DataFrames

The pd.concat() function is used to concatenate two or more DataFrames along a particular axis. By default, axis=0 concatenates the DataFrames vertically (row-wise), while axis=1 concatenates horizontally (column-wise).

```python
import pandas as pd

# Sample data
df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Age': [30, 25]
})

df2 = pd.DataFrame({
    'Name': ['Charlie', 'David'],
    'Age': [35, 40]
})

# Concatenate vertically (row-wise)
concat = pd.concat([df1, df2], axis=0)
print(concat)
```

# **11. Working with Time Series Data**

1-Create a Time Index

The pd.date_range() function generates a range of dates with a specified start date, end date, and frequency. In your case, the frequency is set to 'D', which means daily frequency.

```python
import pandas as pd

# Create a range of dates from '2023-01-01' to '2023-01-10' with daily frequency
date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
print(date_rng)
```

2-Set Index

The df.set_index() function sets the specified column ('Date' in this case) as the DataFrame's index. The inplace=True parameter modifies the original DataFrame directly.

```python
# Create a DataFrame with some data
data = {'Value': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55]}
df = pd.DataFrame(data)

# Assign the date range as the 'Date' column
df['Date'] = date_rng

# Set 'Date' column as the index
df.set_index('Date', inplace=True)
print(df)
```

3-Resample

The df.resample() function is used to resample the time series data at a specified frequency. In your example, the frequency '3D' means you want to resample the data into 3-day periods, summing the Value for each period.

```python
# Resample the data to 3-day periods and sum the values within each period
resampled = df.resample('3D').sum()
print(resampled)
```

Key Points:

Creating a Time Index: pd.date_range() is used to generate a range of dates with a specified frequency.

Setting an Index: The set_index() function assigns the time series column as the DataFrame's index.

Resampling: The resample() function allows you to group data into different time periods. The argument '3D' means 3-day intervals, and .sum() aggregates the values for each interval.


Resampling Frequency Options:

'D': Day

'W': Week

'M': Month

'Q': Quarter

'A': Year

'3D': 3-day intervals (or any other custom frequency)

# **12. Visualization with Pandas**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {'Salary': [45000, 47000, 49000, 51000, 53000, 55000, 57000, 59000, 61000, 63000]}
date_rng = pd.date_range(start='2023-01-01', periods=10, freq='D')
df = pd.DataFrame(data, index=date_rng)

# Plot the 'Salary' column over time
df['Salary'].plot(kind='line', title='Salary Over Time')

# Show the plot
plt.show()
```
